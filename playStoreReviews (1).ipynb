```python
from google_play_scraper import Sort, reviews_all
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display
from google.colab import files

# Create input widgets
app_id_widget = widgets.Text(description='App ID:')
language_widget = widgets.Text(description='Language:', value='en')
country_widget = widgets.Text(description='Country:', value='us')

# Create drop-down for Sort Option
sort_options = ['most_relevant', 'newest']
sort_option_widget = widgets.Dropdown(description='Sort Option:', options=sort_options)

score_filter_widget = widgets.Text(description='Score Filter:')
num_entries_widget = widgets.IntText(description='Number of Entries:', value=10)

# Create GUI button
extract_button = widgets.Button(description='Extract Reviews')

# Display input widgets and button
input_widgets = widgets.VBox([app_id_widget, language_widget, country_widget,
                              sort_option_widget, score_filter_widget, num_entries_widget])
gui = widgets.VBox([input_widgets, extract_button])
display(gui)

def extract_reviews(button):
    # Get user input values
    app_id = app_id_widget.value
    language = language_widget.value
    country = country_widget.value
    sort_option = sort_option_widget.value
    score_filter = score_filter_widget.value

    # Determine the sorting option
    if sort_option.lower() == 'newest':
        sort = Sort.NEWEST
    else:
        sort = Sort.MOST_RELEVANT

    # Convert score_filter to int if provided
    if score_filter:
        score_filter = int(score_filter)

    # Get reviews using the provided parameters
    result = reviews_all(
        app_id,
        sleep_milliseconds=0,
        lang=language,
        country=country,
        sort=sort,
        filter_score_with=score_filter
    )

    # Convert the review data to a DataFrame
    scrapeddata = pd.DataFrame(np.array(result), columns=['review'])
    scrapeddata = scrapeddata.join(pd.DataFrame(scrapeddata.pop('review').tolist()))

    # Get the desired number of entries
    num_entries = num_entries_widget.value
    extracted_data = scrapeddata.head(num_entries)

    # Display the extracted data
    display(extracted_data.head())

    # Save the remaining data to a CSV file
    remaining_data = scrapeddata.iloc[num_entries:]
    remaining_filename = "remaining_reviews.csv"
    remaining_data.to_csv(remaining_filename, index=False)
    print("Remaining data saved to '{}".format(remaining_filename))

    # Save the extracted data to a CSV file
    extracted_filename = "extracted_reviews.csv"
    extracted_data.to_csv(extracted_filename, index=False)
    print("Extracted data saved to '{}".format(extracted_filename))

    # Manually trigger file download
    files.download(extracted_filename)

# Link the button to the function
extract_button.on_click(extract_reviews)
```
